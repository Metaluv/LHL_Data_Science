## Matplotlib 3D Assignment

```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

------------

#### Task 1: Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>

* hyperbolic paraboloid equation is: z = x*x - y*y

```python
# wireframe
def f(x, y):
    return (x ** 2 - y ** 2)

x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('wireframe');
```


-------------
#### Task 2: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* eliptical paraboloid : z = x*x + y*y

```python
# surface plot

def f(x, y):
    return (x ** 2 + y ** 2)

x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='cool')
ax.set_title('surface');
```

#### Task 3: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* Also create a contour projection on the plane z=0 (HINT: use the parameter 'zdir' of contour plot)

* eliptical paraboloid : z = x*x + y*y

```python
# surface plot

def f(x, y):
    return (x ** 2 + y ** 2)

x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z,50, cmap='cool', zdir='z')

cset = ax.contour(X, Y, Z, zdir='z', offset=0, cmap='cool')
ax.set_title('surface');
```
