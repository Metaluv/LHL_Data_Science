# Matplotlib Intro

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plot

```python
x = np.linspace(0,100,200) # 200 evenly spaced numbers between [0,100]
y = np.sin(x)

plt.plot(x,y,'red')
plt.show()
```

## Scatter Plot

```python
# 200 random values [0,10]
x = 10*np.random.rand(200,1)

y = 15*np.random.rand(200,1)

plt.scatter(x,y)
plt.show()
```

## Histogram

```python
y = 15*np.random.rand(200,1)

plt.hist(y,bins=20)
plt.show()
```

## Graphs on Common Axes

```python
x = np.linspace(0,7,200)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = (np.sin(x))*(np.cos(x))


plt.plot(x,y1,'red')
plt.plot(x, y2, 'blue')
plt.plot(x, y3, 'green')


plt.show()
```

## Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y1,'red')
plt.title('sin(x)')

# Second figure
plt.subplot(2,2,2)
plt.plot(x,y2,'blue')
plt.title('cos(x)')

plt.subplot(2,2,3)
plt.plot(x,y3,'blue')
plt.title('3')

plt.subplot(2,2,4)
plt.plot(x,y2,'blue')
plt.title('4')

plt.tight_layout()
plt.show()
```

## Legends

```python
import pandas as pd
from sklearn import datasets
```

```python
# load iris dataset
iris = datasets.load_iris()

# create dataframe
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# create target
iris_df['target'] = iris.target

# map the target values to the target names
iris_df['target_name'] =iris_df.target.map(
    {0: 'setosa', 
     1: 'versicolor',
     2: 'virginica'}
)
```

```python
iris_df.head()
```

```python
# Iris setosa
setosa = iris_df[iris_df.target_name == 'setosa']

# Iris versicolor
versicolor = iris_df[iris_df.target_name == 'versicolor']

# Iris virginica
virginica = iris_df[iris_df.target_name == 'virginica']

# plot setosa
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'],
            marker ='o', color = 'red', label = 'setosa')

# plot versicolor
plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'],
            marker ='o', color = 'green', label = 'versicolor')

# plot virginica
plt.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'],
            marker ='o', color = 'blue', label = 'virginica')

# legend location
plt.legend(loc='upper right')

# plot title
plt.title('Iris flower')

# x-axis title
plt.xlabel('sepal length (cm)')

# y-axis title
plt.ylabel('sepal width (cm)')
plt.show()
```

## Annotations

```python
# the same code as before
plt.scatter(setosa['sepal length (cm)'],setosa['sepal width (cm)'],
            marker ='o', color = 'red', label = 'setosa')

plt.scatter(versicolor['sepal length (cm)'],versicolor['sepal width (cm)'],
            marker ='o', color = 'green', label = 'versicolor')

plt.scatter(virginica['sepal length (cm)'],virginica['sepal width (cm)'],
            marker ='o', color = 'blue', label = 'virginica')

# new lines of code
# it can be tricky to find the right coordinates for the first time
######################
plt.annotate('setosa', xy =(5.0,3.5),
             xytext = (4.25,4.0), arrowprops={'color':'red'})
plt.annotate('virginica', xy =(7.2,3.6),
             xytext = (6.5,4.0), arrowprops={'color':'red'})
plt.annotate('versicolor', xy =(5.05,1.95),
             xytext = (5.5,1.75), arrowprops={'color':'red'})
######################

# the same code as before
plt.legend(loc='upper right')
plt.title('Iris flower')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.ylim(1.5,4.7)
plt.show()
```
