## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../Day_1/dataviz_exercises-master/company_sales_data.csv')
```

```python
df.head()
```

### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
plt.plot(df.month_number,df.total_profit)
plt.title('Total profit per month')
plt.show()
```

```python
# stateless approach
fig, ax = plt.subplots(figsize= (10,6))

ax.plot(df.month_number, df.total_profit)
ax.set_title('Total Profit')
plt.show()
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
x = df.month_number

product_names = [x for x in df.columns][1:-2]
for product in product_names:
    plt.plot(x,df[product], label=product)
plt.legend()
plt.show()
```

```python
# statelss approach

products = df.columns[1:-2]
#print(products)

fig, ax = plt.subplots(figsize=(10,6))

for i in range(len(products)):
    ax.plot(df.month_number, df[products[i]], label = products[i])

ax.legend()            
plt.show()
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
plt.hist(df.total_profit,bins=5)
plt.show()
```

```python
# Stateless approach
fig, ax = plt.subplots(figsize=(10,6))

ax.hist(df.total_profit, bins=10)
plt.show()
```
