# Matplotplib Stateless approach


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

## Line Plot

```python
x = np.linspace(0,100,200)
y = np.sin(x)

fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(x, y, 'red')
plt.show()
```

## Scatter Plot

```python
x = 10*np.random.rand(200,1)
y = 15*np.random.rand(200,1)

fig, ax = plt.subplots(figsize = (10, 6))

ax.scatter(x,y)
plt.show()
```

## Histogram

```python
y = 15 * np.random.rand(200,1)

fig, ax = plt.subplots(figsize = (10,6))

ax.hist(y, bins=20)
plt.show()
```

## Graphs on common axis

```python
x = np.linspace(0, 100, 200)
y1 = np.sin(x)
y2 = (np.sin(x))*(np.cos(x))

fig, ax = plt.subplots(figsize= (10, 6))

ax.plot(x, y1, 'red')
ax.plot(x, y2, 'blue')
plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (10,6))

axes[0].plot(x, y1, 'red')
axes[0].set_title('sin(x)')

axes[1].plot(x, y2, 'blue')
axes[1].set_title('sin(x) * cos(x)')

fig.tight_layout()
plt.show()
```

## Legends

```python
iris = datasets.load_iris()

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

iris_df['target'] = iris.target

iris_df['target_name'] = iris_df.target.map(
    {0: 'setosa',
     1: 'versicolor',
     2: 'virginica'}
)
```

```python
iris_df.tail()
```

```python
# Iris setosa
setosa = iris_df[iris_df.target_name == 'setosa']

# Iris versicolor
versicolor = iris_df[iris_df.target_name == 'versicolor']

# Iris virginica
virginica = iris_df[iris_df.target_name == 'virginica']

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize=(10,6))

# scatter the plot of setosa
ax.scatter(setosa['sepal length (cm)'],setosa['sepal width (cm)'],
            marker ='o', color = 'red', label = 'setosa')


# scatter the plot of versicolor
ax.scatter(versicolor['sepal length (cm)'],versicolor['sepal width (cm)'],
            marker ='o', color = 'green', label = 'versicolor')


# scatter the plot of virginica
ax.scatter(virginica['sepal length (cm)'],virginica['sepal width (cm)'],
            marker ='o', color = 'blue', label = 'virginica')

# set the legend
ax.legend(loc='upper left')

# set the title
ax.set_title('Iris flower')

# set the x-label
ax.set_xlabel('sepal length (cm)')

# set the y-label
ax.set_ylabel('sepal width (cm)')
plt.show()
```

## Annotations

```python
# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize=(10,6))

# scatter the plot of setosa
ax.scatter(setosa['sepal length (cm)'],setosa['sepal width (cm)'],
            marker ='o', color = 'red', label = 'setosa')


# scatter the plot of versicolor
ax.scatter(versicolor['sepal length (cm)'],versicolor['sepal width (cm)'],
            marker ='o', color = 'green', label = 'versicolor')


# scatter the plot of virginica
ax.scatter(virginica['sepal length (cm)'],virginica['sepal width (cm)'],
            marker ='o', color = 'blue', label = 'virginica')

# new lines of code
######################
ax.annotate('setosa', xy =(5.0,3.5),
             xytext = (4.25,4.0), arrowprops={'color':'red'})
ax.annotate('versicolor', xy =(7.2,3.6),
             xytext = (6.5,4.0), arrowprops={'color':'red'})
ax.annotate('virginica', xy =(5.05,1.95),
             xytext = (5.5,1.75), arrowprops={'color':'red'})
######################


# set the legend
ax.legend(loc='upper right')

# set the title
ax.set_title('Iris flower')

# set the x-label
ax.set_xlabel('sepal length (cm)')

# set the y-label
ax.set_ylabel('sepal width (cm)')

# set the x-range 
ax.set_xlim(4,8)

# set the y-range
ax.set_ylim(1.5,4.7)
plt.show()
```
