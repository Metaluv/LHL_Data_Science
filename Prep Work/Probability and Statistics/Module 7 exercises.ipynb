Conditional Probability: Soccer Game Example


```python
# how many times we run the simulation
n_runs = 100000

# initialize variables
SAM = 0
SAM_YES = 0
SAM_NO = 0
ALEX = 0
ALEX_YES = 0
ALEX_NO = 0
YES = 0
NO = 0
```

```python
# process simulation
for _ in range(n_runs):
    # Sam (probability of Sam being the coach)
    if np.random.random() < 0.6:
        SAM += 1

        # YES
        if np.random.random() < 0.5:
            SAM_YES += 1
            YES +=1

        # NO
        else:
            SAM_NO += 1
            NO += 1

    #Alex (probability of Alex being the coach)
    else:
        ALEX += 1

        # YES
        if np.random.random() < 0.3:
            ALEX_YES += 1
            YES += 1

        # NO
        else:
            ALEX_NO += 1
            NO +=1
```

```python
# create probabilities
P_SAM = SAM / n_runs * 100
P_SAM_YES = SAM_YES / n_runs * 100
P_SAM_NO = SAM_NO / n_runs * 100
P_ALEX = ALEX / n_runs * 100
P_ALEX_YES = ALEX_YES / n_runs * 100
P_ALEX_NO = ALEX_NO / n_runs * 100
P_YES = YES / n_runs * 100
P_NO = NO / n_runs * 100
```

```python
print(f'Sam is the coach: {P_SAM}%')
print(f'Sam is the coach and you are the goalkeeper: {P_SAM_YES}%')
print(f'Sam is the coach and you are not the goalkeeper: {P_SAM_NO}%')
print(f'Alex is the coach: {P_ALEX}%')
print(f'Alex is the coach and you are the goalkeeper: {P_ALEX_YES}%')
print(f'Alex is the coach and you are not the goalkeeper: {P_ALEX_NO}%')
print(f'You are the goalkeeper: {P_YES}%')
print(f'You are not the goalkeeper: {P_NO}%')
```

```python
# P(Sam | goalkeeper)
P_SAM_GOAL = P_SAM_YES / P_YES
print(P_SAM_GOAL)
```

Law of Large Numbers Example

```python
# set a random seed to replicate the results
np.random.seed(42)

# generate ten random numbers (0 or 1) with equal probabilities
coin_flips_10 = np.random.randint(0,2,10)
# how many heads in 10 coin flips
count_heads = sum(coin_flips_10 == 1)
print(count_heads)
```

```python
# empty list used to store the results
heads_ratio_nflips = []

# generate integers from 5 to 10,000
n_flips = np.arange(5,10000)

for flips in n_flips:
    # how many heads / flips
    heads_ratio = sum(np.random.randint(0,2,flips) == 1) / flips

    # append ratios
    heads_ratio_nflips.append(heads_ratio)
```

```python
# plot settings
plt.figure(figsize=(10,8))
plt.figure(figsize=(10,8))
plt.xlabel('Flips')
plt.ylabel('Heads ratio')

# number of flips on the x axe and heads ratio on y axe
plt.plot(n_flips, heads_ratio_nflips)

# expected ratio
plt.plot(n_flips, len(n_flips)*[0.5], 'r--')

plt.show()
```

Hypothesis Testing Example

```python
# set a random seed to replicate the results
np.random.seed(42)

# sales history in days
history = 365

# generate one-year sales for store A
mean_A = 20
std_A = 5
shop_A_sales = np.random.normal(mean_A, std_A, history)

# generate one-year sales for store B
mean_B = 19.5
std_B = 5
shop_B_sales = np.random.normal(mean_B, std_B, history)
```

```python
# H0 = the mean of sales of shop A equals to the mean of sales of shop B
# HA = the means are not equal

# set the significance level
alpha = 0.05

# print the store A mean
print(shop_A_sales.mean())

# print the store B mean
print(shop_B_sales.mean())

# the differnce in the means
observed_means_diff = shop_A_sales.mean() - shop_B_sales.mean()
print(observed_means_diff)
```

```python
# simulate what it would look like if both stores' sales were identically distributed
both_sales = np.concatenate((shop_A_sales, shop_B_sales))

# permutation (random re-ordering of an array)
sales_perm = np.random.permutation(both_sales)

# permutation replicates
perm_shop_A_sales = sales_perm[:len(shop_A_sales)]
perm_shop_B_sales = sales_perm[len(shop_A_sales):]

print(perm_shop_A_sales.mean() - perm_shop_B_sales.mean())
```

```python
# create an empty list to store the permutation replicates means
perm_repl_means = []

for _ in range(1000):
    # permutation
    sales_perm = np.random.permutation(both_sales)

    # permutation replicates
    perm_shop_A_sales = sales_perm[:len(shop_A_sales)]
    perm_shop_B_sales = sales_perm[len(shop_A_sales):]

    # permutation replicates mean
    perm_repl_mean = perm_shop_A_sales.mean() - perm_shop_B_sales.mean()

    # append perm_repl_mean to list
    perm_repl_means.append(perm_repl_mean)
```

```python
# compute the p-value
p = np.sum(np.abs(perm_repl_means) >= observed_means_diff) / len(perm_repl_means)

# print the result
print('p-value =', p)
```

```python
# final decision
if p < alpha:
    print('H0 is rejected.')
else:
    print('H0 is not rejected.')
```
